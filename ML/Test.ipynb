```python
import numpy as np

np1 = np.array([3, 6, 0, 1, 2, 3, 4, 5])

x = np.where(np1 == 3) # 4

print(x) # (array([0, 5]),)

print(x[0]) # [0 5]

print(np1[x[0]]) # [3 3]

y = np.where(np1 % 2 == 0) # 4

print(y) # (array([1, 2, 4, 6]),)
```

```python
import numpy as np

np1 = np.array([3, 6, 0, 1, 2, 3, 4, 5])

filtered = []

for thing in np1:
	if thing % 2 == 0:
		filtered.append(True)
	else:
		filtered.append(False)

print(np1[filtered])

filtered2 = np1 % 2 == 0
print(np1[filtered2])
```
